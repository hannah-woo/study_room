```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.title('sigmoid function')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 =  sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(1.5*x)

plt.plot(x, y1, color='red')
plt.plot(x, y2, color='green')
plt.plot(x, y3, color='blue')
plt.title('sigmoid function')
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)
plt.plot(x, y1, color='red')
plt.plot(x, y2, color='green')
plt.plot(x, y3, color='blue')
plt.title('sigmoid function')
plt.show()
```

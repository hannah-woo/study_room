### seaborn 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
plt.rc('font', family='Malgun Gothic')
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(data=tips, x='day')
plt.title('요일별 팁을 준 횟수')
plt.show()
```

```python
sns.boxplot(data=tips, x='day', y='tip')
plt.title('요일별 팁의 박스 플롯')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('식사 금액과 팁의 관계')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.suptitle('식사 금액과 팁의 관계 joint plot', y=1.05)
```

```python
sns.boxplot(data=tips, x='day', y='tip', hue='sex')
plt.title('요일별, 성별 팁의 박스 플롯')
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='tip', hue='sex')
plt.title('요일별, 성별 팁의 바이올린 플롯')
plt.show()
```

```python
sns.displot(tips.tip, kde=True, rug=True, height=3, aspect=2)
plt.title('팁의 분포')
plt.show()
```

```python
sns.displot(tips, x='day', y='total_bill')
plt.title('식사 금액의 분포')
plt.show()
```
